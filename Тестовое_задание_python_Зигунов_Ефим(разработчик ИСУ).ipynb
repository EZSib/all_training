```python
data = {'Отдел': [1,1,2,3,1,3], 'Сотрудник': [1,2,3,6,7,6]}
df = pd.DataFrame(data)
```

```python
uniq_emp = df.groupby('Отдел')['Сотрудник'].nunique()
df['cnt'] = df['Отдел'].apply(lambda x: uniq_emp[x])
```

```python
df1 = pd.DataFrame({'w':[1,1,2,2,2,3,3,3,4,4,4],
                   'user':['User2', 'User3','User1','User1','User2','User2','User2','User3','User1','User1','User3'],
                   'c':[7,2,3,1,2,5,4,6,9,8,7]})
```

Не нашёл в пандасе метода, который бы смог, как sql агригрировать датафрейм, и подставить необходимые значения поэтому напишу своё решение.

```python
# создание датафрейма нужного вида
d= []
for j in df1['w'].unique():
  for i in sorted(df1['user'].unique()):
    d.append({'w': j, 'user': i})

df2 = pd.DataFrame(d)
```

```python
# демонстрация того, как будут подставляться значения: по ключам неделя и юзер нахожу посещения, если ключа нет - 0
dict(df1.groupby(['w','user'])['c'].sum()).get((1,'User2'),0)
```

```python
def find_w_user(w_usr):
  list_cnt = []
  for i in w_usr:
        list_cnt.append(dict(df1.groupby(['w','user'])['c'].sum()).get((i[0],i[1]),0))
  return(list_cnt)
```

```python
df2['cnt'] = find_w_user(list(df2[['w','user']].values))
```

```python
df2['sum'] = df2.groupby(['user'])['cnt'].cumsum()
```
